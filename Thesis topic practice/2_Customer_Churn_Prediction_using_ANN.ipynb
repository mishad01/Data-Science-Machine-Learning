```python
df.drop(columns= ['RowNumber','CustomerId','Surname'],inplace = True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
#ALTERNATIVE
#geography = pd.get_dummies(X['Geography'], drop_first=True)
#gender = pd.get_dummies(X['Gender'], drop_first=True)
```

```python
X = df.drop(columns=['Exited'])
y = df['Exited'].values

#X = dataset.iloc[:,3:13];
#y = dataset.iloc[:,13];
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler  # Importing StandardScaler for feature scaling
sc = StandardScaler()  # Creating an instance of StandardScaler
X_train_scaled = sc.fit_transform(X_train)  # Fit the scaler to the training data and scale it
X_test_scaled  = sc.transform(X_test)  # Apply the same scaling to the test data (without fitting again)
```

```python
X_train_scaled
```

```python

import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
#model = Sequential()  # Initialize a Sequential Neural Network

# Add the input layer with 3 neurons, sigmoid activation, and 11 input features
#model.add(Dense(3, activation='sigmoid', input_dim=11))  # Input layer with 11 features (input_dim)

# Add the output layer with 1 neuron (binary classification) and sigmoid activation
#model.add(Dense(1, activation='sigmoid'))  # Output layer with 1 neuron for binary output (0 or 1)
```

```python
model = Sequential()
model.add(Dense(11,activation='relu',input_dim=11))
model.add(Dense(11,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
#model.compile(loss='binary_crossentropy',optimizer='Adam')
```

```python
model.compile(loss='binary_crossentropy',optimizer = 'Adam',metrics=['accuracy'])
```

```python
#model.fit(X_train_scaled,y_train,epochs=10)
```

```python
# Train the model on the training data
model.fit(
    X_train_scaled,          # Input features after scaling
    y_train,                 # Target labels for training
    epochs=100,              # Number of times the model will pass through the entire training dataset
    validation_split=0.2     # Use 20% of the training data for validation
)
```

```python
model.layers[0].get_weights()
```

```python
model.predict(X_test_scaled)
```

```python
y_log = model.predict(X_test_scaled)
```

```python
y_pred = np.where(y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

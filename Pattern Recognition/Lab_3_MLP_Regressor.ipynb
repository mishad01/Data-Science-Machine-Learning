```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Patter Recognition Lab Dataset /song_data.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['song_name'] = label_encoder.fit_transform(df['song_name'])
```

```python
df.head()
```

```python
X = df.iloc[:,[2,3,4,5,6,7]].values
y = df.iloc[:,1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

#Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neural_network import MLPRegressor
regressor = MLPRegressor(hidden_layer_sizes=(64,64),activation = 'relu',random_state=0,max_iter=2000).fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
